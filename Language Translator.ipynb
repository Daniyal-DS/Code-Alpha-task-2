```python
import asyncio
from googletrans import Translator,LANGUAGES
import ipywidgets as widgets
from IPython.display import display
```

```python
async def translator_function(from_lan, to_lan, t_text)->str:
    Tt=""
    try:
        async with Translator() as translator:
             translated_text = await translator.translate( t_text, src=from_lan, dest=to_lan)
             Tt= translated_text.text
    except:
        Tt= "Cannot translate"
    return f"{Tt}."
```

```python
_, options = zip(*LANGUAGES.items())

from_dropdown = widgets.Dropdown(
    options=options,
    description='From:',
)


to_dropdown = widgets.Dropdown(
    options=options,
    description='To:',
)


text_editor = widgets.Textarea(
    description='Write text:',
    layout=widgets.Layout(width='90%')
)

output_text_box = widgets.Text(
    description='Translated:',
    layout=widgets.Layout(width='90%')
)


def display_text(change):
    frm_l = next((k for k, v in LANGUAGES.items() if v == from_dropdown.value), None)
    to_l = next((k for k, v in LANGUAGES.items() if v == to_dropdown.value), None)
    translated_text=""
    input_text = text_editor.value.strip()

    async def run_translation():
        translated_text = await translator_function(frm_l, to_l, input_text)
        output_text_box.value = translated_text

    asyncio.create_task(run_translation())  # Schedule the async function


button = widgets.Button(description="Translate")
button.on_click(display_text)


display(
    widgets.VBox([
        widgets.Label("Select the languages first"),
        from_dropdown,
        to_dropdown,
        text_editor,
        button,
        output_text_box
    ])
)
```
